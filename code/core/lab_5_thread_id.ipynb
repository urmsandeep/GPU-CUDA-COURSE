```python
%%writefile thread_id.cu

#include <stdio.h>

__global__ void printThreadInfo() {
    // Calculate global thread ID
    int tid = blockIdx.x * blockDim.x + threadIdx.x;

    // Print block ID, thread ID, and global ID
    printf("Block %d, Thread %d, Global ID: %d\n",
           blockIdx.x, threadIdx.x, tid);
}

int main() {
    // Launch with 4 blocks, 8 threads each
    printThreadInfo<<<4, 8>>>();
    cudaDeviceSynchronize();
    return 0;
}
```
